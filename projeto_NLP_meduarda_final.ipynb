```python
import nltk
from nltk.tokenize import word_tokenize
# tokenizador de sentença do nltk
nltk.download('punkt_tab')

# tokenized_sents = [word_tokenize(i, 'portuguese') for i in avaliacao_tablet['content'].values]
```

```python
def tokenize_df(list_of_elements):
  tokenized_sents = []
  words_in_total = []
  max_len_sent = 0

  for content in list_of_elements:
    tokenized_phase = word_tokenize(content, 'portuguese')
    for word in tokenized_phase:
      words_in_total.append(word)

    if len(tokenized_phase) > max_len_sent:
      max_len_sent = len(tokenized_phase)
    tokenized_sents.append(tokenized_phase)
  # word_in_total_voc = set(words_in_total)
  return tokenized_sents, max_len_sent, words_in_total
```

```python
def resultado_tokenizacao_x_primeiras_frases(qnt_frases,tokenized_sents, words_in_total, max_len_sent):
  print(f'Apresentação de como ficam as {qnt_frases} primeiras avaliações Tokenizadas:\n')
  for i,list_sent in enumerate(tokenized_sents):
    if i < qnt_frases:
      print(i,list_sent)
    else:
      break
```

```python
def parametros_finais_tokenizacao(words_in_total, max_len_sent):
  print(f'\nMaior Vetor de palavras: {max_len_sent}')
  print(f'Total Tokens: {len(words_in_total)}')
  print(f'Tamanho do vocabulário: {len(set(words_in_total))}')
```

##Dataloader e separação do dataset

```python
class TextDataset(Dataset):
  def __init__(self, X, y, device):
    self.device = device
    self.X = X
    self.X = self.X.to(device)
    self.y = y
    self.y = self.y.to(device)
    self.len = len(y)

  def __len__(self):
    return self.len

  def __getitem__(self, idx):
    input_ids = (self.X[0, idx]).to(device)  # X[0, :] é o input_ids
    attention_mask = (self.X[1, idx]).to(device)  # X[1, :] é o attention_mask
    label = self.y[idx].to(device)
    return input_ids, attention_mask, label
    # return self.X[:,idx], self.y[idx]
```

```python
dataset_bert = TextDataset(X_bert, Y_bert, device)
```

```python
#Calculo do número de instâncias
num_train_inst = np.int64(np.round(dataset_bert.len*train_ratio))
num_val_inst = np.int64(np.round(dataset_bert.len*validation_ratio))
num_test_inst = np.int64(np.round(dataset_bert.len*test_ratio))
```

##Treinamento e Validação

```python
epoch = 30
steps_per_epoch = 100
epoch_validation_samples = 50
num_classes = 5
```

```python
model_bert = BertForSequenceClassification.from_pretrained('neuralmind/bert-base-portuguese-cased').to(device)
for param in model_bert.bert.parameters():
  param.requires_grad = False

model_bert.classifier = torch.nn.Linear(model_bert.classifier.in_features, num_classes).to(device)

loss_func = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model_bert.parameters())

acc_calc = lambda output, labels: (labels==output.argmax(axis=1)).sum()
```

```python
num_classes = model_bert.classifier.out_features
print(f'O número de classes é: {num_classes}')
```

```python
for batch_idx, (input_ids, input_masks, labels) in enumerate(train_bert_loader):
    print(f"Batch {batch_idx + 1} - Input shape: {input_ids.shape};Input masks: {input_masks.shape};  Labels shape: {labels.shape}")
    break
```

```python
print(device)
```

```python
epoch_metadata = []

for i in range(epoch):
  num_train_examples = 0
  num_val_examples = 0

  train_hits = 0
  val_hits = 0

  train_bar = tqdm(total=steps_per_epoch, desc='Train',unit='steps',position=0,leave=True)
  val_bar = tqdm(total=epoch_validation_samples, desc='val',unit='samples',position=0,leave=True)

  for batch_number, (input_ids, input_masks,labels) in enumerate(train_bert_loader):
    train_running_loss = 0
    model_bert.train()

    input_ids = input_ids.to(device)
    input_masks = input_masks.to(device)
    labels = labels.to(device)

    # Passar pelos dados
    logits = model_bert(input_ids.to(device), attention_mask=input_masks.to(device)).logits
    logits = logits.to(device)

    # Verificar as dimensões
    # print(f"Logits shape: {logits.shape}")  # Esperado: (batch_size, num_classes)
    # print(f"Labels shape: {labels.shape}")  # Esperado: (batch_size,)

    # input_ids, input_masks = features[:,0,:], features[:,1,:]
    loss = loss_func(logits, labels.long())


    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    train_running_loss += loss.item()

    softmax_pedictions = torch.nn.functional.softmax(logits,dim=1)
    train_hits += acc_calc(softmax_pedictions,labels)

    train_bar.update(1)
    num_train_examples += input_ids.shape[0]


    if (batch_number+1) % steps_per_epoch == 0:
      train_bar.close()
      break

  for batch_number, (input_ids, input_masks,labels) in enumerate(val_bert_loader):
    with torch.no_grad():
      val_running_loss = 0
      model_bert.eval()


      input_ids = input_ids.to(device)
      input_masks = input_masks.to(device)
      labels = labels.to(device)

      # Passar pelos dados
      logits = model_bert(input_ids, attention_mask=input_masks).logits
      logits = logits.to(device)
      loss = loss_func(logits, labels.long())

      val_running_loss += loss.item()

      softmax_pedictions = torch.nn.functional.softmax(logits,dim=1)
      val_hits += acc_calc(softmax_pedictions,labels)

      val_bar.update(1)
      num_val_examples += input_ids.shape[0]


      if (batch_number+1) % epoch_validation_samples == 0:
        val_bar.close()
        break
  train_acc = torch.true_divide(train_hits, num_train_examples)
  val_acc = torch.true_divide(val_hits, num_val_examples)

  print(f'epoch {i+1}:Train loss: {train_running_loss}; Train acc: {train_acc}; Val loss: {val_running_loss}; Val acc: {val_acc}')
```
